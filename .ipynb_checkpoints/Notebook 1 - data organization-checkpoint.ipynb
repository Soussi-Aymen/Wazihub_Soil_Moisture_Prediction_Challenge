### Separate Train in 4 dataframes corresponding to each field

```python
# rename column names in Train
mapping = {"Soil humidity 1":"Humidity1", "Soil humidity 2":"Humidity2", 
           "Soil humidity 3":"Humidity3", "Soil humidity 4":"Humidity4",
           "Irrigation field 1":"Irrigation1", "Irrigation field 2":"Irrigation2",
           "Irrigation field 3":"Irrigation3", "Irrigation field 4":"Irrigation4",
           "Air temperature (C)":"Air_Temp", "Air humidity (%)":"Air_humidity",
           "Pressure (KPa)":"Pressure", "Wind speed (Km/h)":"Wind_Speed",
           "Wind gust (Km/h)":"Wind_Gust", "Wind direction (Deg)":"Wind_Direction"}
Train.rename(mapper=mapping, axis=1, inplace=True)
```

```python
# separate fields
Train1 = Train[["timestamp", "Humidity1", "Irrigation1", "Air_Temp", "Air_humidity", 
                "Pressure", "Wind_Speed", "Wind_Gust", "Wind_Direction"]]
Train2 = Train[["timestamp", "Humidity2", "Irrigation2", "Air_Temp", "Air_humidity", 
                "Pressure", "Wind_Speed", "Wind_Gust", "Wind_Direction"]]
Train3 = Train[["timestamp", "Humidity3", "Irrigation3", "Air_Temp", "Air_humidity", 
                "Pressure", "Wind_Speed", "Wind_Gust", "Wind_Direction"]]
Train4 = Train[["timestamp", "Humidity4", "Irrigation4", "Air_Temp", "Air_humidity", 
                "Pressure", "Wind_Speed", "Wind_Gust", "Wind_Direction"]]
```

```python
# remove excess data
Train1 = Train1[np.isfinite(Train1["Irrigation1"])]
Train2 = Train2[np.isfinite(Train2["Irrigation2"])]
Train3 = Train3[np.isfinite(Train3["Irrigation3"])]
Train4 = Train4[np.isfinite(Train4["Irrigation4"])]

# print new shapes
print("Shape of Train :", Train.shape)
print("Shape of Train1:", Train1.shape)
print("Shape of Train2:", Train2.shape)
print("Shape of Train3:", Train3.shape)
print("Shape of Train4:", Train4.shape)
```

### Train data limits - test data

```python
# dataframe with train/test data limits - get train start/end timestamps
limits = pd.DataFrame(index=["train_start", "train_end", "test_start", "test_end"],
                      columns=["field1", "field2", "field3", "field4"])
fields = ["field1", "field2", "field3", "field4"]
trains = [Train1, Train2, Train3, Train4]
for field, train in zip(fields, trains):
    limits[field]["train_start"] = train["timestamp"][0]
    index = train[np.isfinite(train["Air_Temp"])].index.max()
    limits[field]["train_end"] = train["timestamp"][index]
```

```python
# extend submission file
def ID2timestamp(s):    #extract timestamp
    split = s.split(" ")
    return split[0] + " " + split[1]

def ID2field(s):        #extract field (ID)
    split = s.split(" ")
    return split[-1]

SampleSubmission["timestamp"] = SampleSubmission["ID"].apply(ID2timestamp)
SampleSubmission["timestamp"] = pd.to_datetime(SampleSubmission["timestamp"], format="%Y-%m-%d %H:%M:%S")
SampleSubmission["field"] = SampleSubmission["ID"].apply(ID2field).astype("int")
```

```python
# get test start/end timestamps
fieldNos = [1, 2, 3, 4]
for field, train, No in zip(fields, trains, fieldNos):
    limits[field]["test_start"] = SampleSubmission["timestamp"][SampleSubmission["field"]==No].min()
    limits[field]["test_end"] = SampleSubmission["timestamp"][SampleSubmission["field"]==No].max()
limits = limits.astype(np.datetime64)
limits
```

### Check for NaN values

```python
# Train1
for col in Train1.columns.values:
    col_str = col + " "*(14-len(col))
    if col in ["timestamp", "Irrigation1"]: 
        missing = Train1[col].isna().sum()
    else: 
        missing = Train1[col][Train1["timestamp"]<=limits["field1"]["train_end"]].isna().sum()
    print("Number of NaN values in " + col_str + ": " + str(missing))
```

```python
# Train2
for col in Train2.columns.values:
    col_str = col + " "*(14-len(col))
    if col in ["timestamp", "Irrigation2"]: 
        missing = Train2[col].isna().sum()
    else: 
        missing = Train2[col][Train2["timestamp"]<=limits["field2"]["train_end"]].isna().sum()
    print("Number of NaN values in " + col_str + ": " + str(missing))
```

```python
# Train3
for col in Train3.columns.values:
    col_str = col + " "*(14-len(col))
    if col in ["timestamp", "Irrigation3"]: 
        missing = Train3[col].isna().sum()
    else: 
        missing = Train3[col][Train3["timestamp"]<=limits["field3"]["train_end"]].isna().sum()
    print("Number of NaN values in " + col_str + ": " + str(missing))
```

```python
# Train4
for col in Train4.columns.values:
    col_str = col + " "*(14-len(col))
    if col in ["timestamp", "Irrigation4"]: 
        missing = Train4[col].isna().sum()
    else: 
        missing = Train4[col][Train4["timestamp"]<=limits["field4"]["train_end"]].isna().sum()
    print("Number of NaN values in " + col_str + ": " + str(missing))
```

##### Reason for NaN values
- From *2019-03-25 22:50:00* to *2019-03-29 22:50:00* it is the 4 day test set for Humidity1: 4\*288 + 1 = 1153
- From *2019-04-19 20:15:00* to *2019-04-23 20:15:00* it is the 4 day test set for Humidity3: 4\*288 + 1 = 1153
- Train3 and Train4 have 2 periods without data, which are test sets for Humidity1 and Humidity3: 1153 + 1153 = 2306
